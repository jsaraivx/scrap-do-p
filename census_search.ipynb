# Census API Variables Search
This notebook downloads the 2024 ACS 1-Year variables table, saves it locally as a CSV, and provides a quick search interface over the `Label` column.

```python
import pandas as pd
import os
import requests
```

```python
URL = 'https://api.census.gov/data/2024/acs/acs1/variables.html'
CSV_FILE = 'census_variables_2024.csv'

if not os.path.exists(CSV_FILE):
    print(f'Downloading data from {URL}...')
    try:
        # The API table can be quite large, we'll use pandas read_html
        # which is robust for basic HTML tables.
        # Setting match='Name' helps identify the right table if there are multiple.
        tables = pd.read_html(URL, match='Name')
        df = tables[0]
        
        # Save to CSV for future fast access
        df.to_csv(CSV_FILE, index=False)
        print(f'Data successfully saved to {CSV_FILE}')
    except Exception as e:
        print(f'Error downloading/parsing data: {e}')
else:
    print(f'Loading data from local file: {CSV_FILE}')
    df = pd.read_csv(CSV_FILE)

# Display basic info
if 'df' in locals():
    print(f'Total variables loaded: {len(df)}')
    display(df.head(3))
```

## 🔍 Interface de Busca
Altere o valor de `termo_busca` abaixo e execute a célula (Shift+Enter) para pesquisar dentro da coluna `Label`.

```python
# ---------------------------------------------------------
# 1. DIGITE SEU TERMO DE BUSCA AQUI
# ---------------------------------------------------------
termo_busca = 'Estimate!!Total:!!Male:!!Under 5 years'

# ---------------------------------------------------------
# 2. EXECUTAR A BUSCA E VER RESULTADOS
# ---------------------------------------------------------
if 'df' in locals() and 'Label' in df.columns:
    # Realiza a busca ignorando maiúsculas/minúsculas (case=False)
    # Preenchemos NaN com '' (string vazia) para evitar erros do .str.contains
    resultados = df[df['Label'].fillna('').str.contains(termo_busca, case=True, regex=False)]
    
    print(f"Encontrados {len(resultados)} resultados para o termo: '{termo_busca}'\n")
    
    # Configura o pandas para não cortar o texto do Label, permitindo leitura completa
    pd.set_option('display.max_colwidth', None)
    display(resultados)
else:
    print("Erro: DataFrame não carregado ou coluna 'Label' ausente.")
```
